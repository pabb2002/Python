# Importing Libraries and Reading Dataframe

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Originaldf=pd.read_csv('owid-energy-data.csv')
```

```python
Originaldf.head()
```

```python
df=Originaldf.dropna()
df
```

# United States Data Over Time

```python
US=df[df['country']=='United States']
```

```python
plt.figure()
plot=sns.lineplot(x='year',y='solar_energy_per_capita',data=US)
plot.set_xlabel('Year')
plot.set_ylabel('Solar Energy Per Capita')
plot.set_title('US Solar Energy Over Time')
plt.xlim(2000,2018)
plt.show()
```

```python
plt.figure()
plot=sns.lineplot(x='year',y='wind_energy_per_capita',data=US)
plot.set_xlabel('Year')
plot.set_ylabel('Wind Energy Per Capita')
plot.set_title('US Wind Energy Over Time')
plt.xlim(2000,2018)
plt.show()
```

```python
plt.figure(figsize=(10,6))

column=['year','solar_consumption','solar_elec_per_capita','solar_energy_per_capita','wind_consumption','wind_elec_per_capita','wind_energy_per_capita']
USEnergyData=US[column].set_index('year')

plot=sns.lineplot(data=USEnergyData)
plot.set_xlabel('Year')
plot.set_title('US Energy Metrics Over Time')
plt.xlim(2000,2018)
plt.show()
```

# Comparing Solar and Wind Energy Data from Two Years for Select Countries

```python
selectCountries=[]

noCountries=int(input('Enter number of countries:'))
FirstYear=int(input('Enter the first year for comparison:'))
SecondYear=int(input('Enter the second year for comparison:'))
```

```python
for i in range(noCountries):
    country=str(input('Enter a country: '))
    selectCountries.append(country)
    
```

```python
selectCountries
```

```python
TwoYears=sortedDataPastDecade.set_index('year').loc[[FirstYear,SecondYear]].reset_index().dropna().drop_duplicates()
TwoYears
```

```python
selectCountriesData=TwoYears[TwoYears['country']==selectCountries[0]]

for i in selectCountries:
    concatDf=TwoYears[TwoYears['country']==i]
    selectCountriesData=pd.concat([selectCountriesData,concatDf])

selectCountriesData=selectCountriesData.reset_index().drop('index',axis=1).drop_duplicates()
selectCountriesData
```

```python
fig,axes=plt.subplots(2,2,figsize=(30,20))
fig.suptitle('Data for Select Countries Comparing Two Years')
sns.barplot(ax=axes[0,0],data=selectCountriesData,x='country',y='solar_energy_per_capita',hue='year')
sns.barplot(ax=axes[0,1],data=selectCountriesData,x='country',y='solar_consumption',hue='year')
sns.barplot(ax=axes[1,0],data=selectCountriesData,x='country',y='wind_energy_per_capita',hue='year')
sns.barplot(ax=axes[1,1],data=selectCountriesData,x='country',y='wind_consumption',hue='year')
```
